```python
manhattan = dataframe["neighbourhood_group"] == "Manhattan"
brooklyn = dataframe["neighbourhood_group"] == "Brooklyn"
queens = dataframe["neighbourhood_group"] == "Queens"
```

```python
dataframe[(manhattan) | (brooklyn) | (queens)]
```

```python
is_mbq = dataframe["neighbourhood_group"].isin(["Manhattan","Brooklyn","Queens"])
dataframe[is_mbq]
```

```python
#identify the top 3 Revenue Generating Neighborhoods within each of the three Neighbourhood_Groups
# use neighbourhood for Harlem, etc.
dataframe[is_mbq].groupby(["neighbourhood_group", "neighbourhood"])["Revenue"].agg([sum, max])

#dataframe[is_mbs].groupby("neighbourhood")["condition to obtain top3 neighborhoods"]

#option A: return .max per neighboourhood
#option B:
#optionC: create condition to obtain average revenue, then filter for top 10%-30%
```

```python
# Need 9 Overall Rows: 3 of the top generating neighbourhoods within each of the 3 Neighbourhood_Groups
top_df = dataframe[is_mbq].groupby(["neighbourhood_group", "neighbourhood"])["Revenue"].agg([sum]).reset_index().sort_values(by='sum',ascending=False)
```

```python
brooklyn_id = top_df[top_df["neighbourhood_group"] =="Brooklyn"].head(3)
manhattan_id = top_df[top_df["neighbourhood_group"] =="Manhattan"].head(3)
queens_id = top_df[top_df["neighbourhood_group"] =="Queens"].head(3)
```

```python
pd.concat([brooklyn_id,manhattan_id,queens_id])
```

```python
pd.concat([brooklyn_id,manhattan_id,queens_id]).plot(x="neighbourhood", kind='bar')
```
